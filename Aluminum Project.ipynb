```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#importing energy for kpoint convergence 
k = np.loadtxt("kpntconv.txt",usecols=(0))
Ek = np.loadtxt("kpntconv.txt",usecols=(1))
```

```python
#Plotting Energy as a function of K
plt.scatter(k,Ek,c='r')
plt.xlabel("number of K points-K^(1/3)-a.u.")
plt.ylabel("Energy-Ry")
plt.title("K point convergence of Al")
print("Converges to ~"+str(Ek[Ek.size-1])+" Ry")
```

```python
#importing files for lattice parameter
v = np.loadtxt("latp.txt",usecols=(0))
v = v**3
Ep = np.loadtxt("latp.txt",usecols=(1))
```

```python
#Plotting Energy As a Function of V
plt.scatter(v,Ep,c='r')
plt.xlabel("Lattice Volume-a^3-(a.u.)^3")
plt.ylabel("Energy-Ry")
plt.title("Minimization in E of Al")
print("minimized at ~"+str(v[int(v.size/2)])+" a.u.^3")
```
